```python
import pandas as pd

df = pd.read_csv('Customer Preference Survey (Responses).csv')
```

```python
df
```

```python
df = df.replace(['ไม่เคยซื้อ'],0)
df = df.replace(['ไม่เคย'],0)
df = df.replace(['ไม่'],0)
df = df.replace(['ไม่เคยซือ'],0)
df = df.replace(['เคยซื้อ'],1)
df = df.replace(['เคย'],1)
df = df.fillna(0)
```

```python
df = df.set_index('Timestamp')
df
```

```python
df.info()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules['consequents_len'] = rules['consequents'].apply(len)
rules['antecedents_len'] = rules['antecedents'].apply(len)
rules["antecedents_str"] = rules["antecedents"].apply(lambda x: list(x)).astype("unicode")
rules["consequents_str"] = rules["consequents"].apply(lambda x: list(x)).astype("unicode")
rules
```

```python
rules1 = rules[(rules['consequents_len'] <= 2) & (rules['antecedents_len'] <= 2)  & (rules['lift'] >= 1) & (rules['confidence'] >= 0.5) & (rules['support'] >= 0.5)].sort_values('lift', ascending=False).head(50)
```

```python
rules2 = rules[(rules['consequents_len'] <= 1) & (rules['antecedents_len'] <= 1)  & (rules['lift'] >= 1) & (rules['confidence'] >= 0.5) & (rules['support'] >= 0.5)].sort_values('lift', ascending=False).head(50)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

fig, ax=plt.subplots(figsize=(10,4))
GA=nx.from_pandas_edgelist(rules1,source='antecedents_str',target='consequents_str')
nx.draw(GA,with_labels=True,font_family='TH Sarabun Chula')
plt.show()
```

```python
item_item_matrix = pd.DataFrame(index=df.columns,columns=df.columns)
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(df.iloc[:,i],df.iloc[:,j])
```

```python
links = item_item_matrix.rename_axis('related item', axis='columns').stack().reset_index()
links.columns = ['item', 'related item','value']
```

```python
links_filtered=links.loc[ (links['value'] > 0.7) & (links['item'] != links['related item']) ]
```

```python
import matplotlib.pyplot as plt
import networkx as nx

fig, ax=plt.subplots(figsize=(15,8))
GA=nx.from_pandas_edgelist(links_filtered,source='item',target='related item')
nx.draw(GA,with_labels=True, font_size =20 , font_family='TH Sarabun Chula')
plt.show()
```
